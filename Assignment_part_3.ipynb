```python
import pandas as pd
```

```python
# df = temp
# df2 = temp, wspd
# df3 = temp, wspd, visib 
# df4 = temp, wdsp, visib, sndp, dewp, slp
df = pd.read_csv('https://raw.githubusercontent.com/16004691da/16004691_DataAnalytics/main/assignmentlineardata-temp.csv', index_col = 0)
df2 = pd.read_csv('https://raw.githubusercontent.com/16004691da/16004691_DataAnalytics/main/assignmentlineardata-temp-wspd.csv', index_col = 0)
df3 = pd.read_csv('https://raw.githubusercontent.com/16004691da/16004691_DataAnalytics/main/assignmentlineardata-temp-wspd-visib.csv', index_col = 0)
df4 = pd.read_csv('https://raw.githubusercontent.com/16004691da/16004691_DataAnalytics/main/assignmentlineardata.csv', index_col = 0)
```

```python
print(df[:6])
```

```python
import numpy as np

shuffle = df.iloc[np.random.permutation(len(df))]

predictors = shuffle.iloc[:,0:1]

print(predictors[:6])
```

```python
shuffle[:5]
```

```python
targets = shuffle.iloc[:, 1]

print(targets[:6])
```

```python
scale_num_offences = 1.0
```

```python
trainsize = int(len(shuffle['num_offences'])*0.8)

testsize = len(shuffle['num_offences']) - trainsize


nppredictors = 1

noutputs = 1
```

```python
%tensorflow_version 1.x
import tensorflow as tf

print(tf.__version__)

import shutil

tf.compat.v1.logging.set_verbosity(tf.compat.v1.logging.INFO)

shutil.rmtree('/tmp/linear_regression_trained_model', ignore_errors=True)

estimator = tf.contrib.learn.SKCompat(tf.contrib.learn.LinearRegressor(model_dir='/tmp/linear_regression_trained_model', optimizer=tf.train.AdamOptimizer(learning_rate=0.1), enable_centered_bias=False, feature_columns=tf.contrib.learn.infer_real_valued_columns_from_input(predictors.values)))

print("starting to train");

estimator.fit(predictors[:trainsize].values, targets[:trainsize].values.reshape(trainsize, noutputs)/scale_num_offences, steps=10000)

preds = estimator.predict(x=predictors[trainsize:].values)

predslistscale = preds['scores'] * scale_num_offences

rmse = np.sqrt(np.mean((targets[trainsize:].values - predslistscale)**2))
print('LinearRegression has RMSE of {0}'.format(rmse));

avg = np.mean(shuffle['num_offences'][:trainsize])

rmse = np.sqrt(np.mean((shuffle['num_offences'][trainsize:] - avg)**2))
print('Just using average = {0} has RMSE of {1}'.format(avg, rmse));
```

```python
input = pd.DataFrame.from_dict(data = 
				{'temp' : [25.5, 18.7, 30.7, 34.9, 46.2, 40.6, 32.6, 34.3, 39.6, 39.2]   
        })
					

estimator = tf.contrib.learn.SKCompat(tf.contrib.learn.LinearRegressor(model_dir='/tmp/linear_regression_trained_model', enable_centered_bias=False, feature_columns=tf.contrib.learn.infer_real_valued_columns_from_input(input.values)))

preds = estimator.predict(x=input.values)

predslistnorm = preds['scores']
predslistscale = preds['scores']*1000
prednorm = format(str(predslistnorm))
pred = format(str(predslistscale))
print(prednorm)
print(pred)

#expected values: 702, 808, 799, 871, 914, 863, 800, 883, 899, 888
```
