```python
scale_num_offences = 1.0
```

```python
trainsize = int(len(shuffle['num_offences'])*0.8)

testsize = len(shuffle['num_offences']) - trainsize


nppredictors = 22

noutputs = 1
```

```python
%tensorflow_version 1.x
import tensorflow as tf


print(tf.__version__)


import shutil  


tf.compat.v1.logging.set_verbosity(tf.compat.v1.logging.INFO)


shutil.rmtree('/tmp/DNN_offences_regression_trained_model', ignore_errors=True)

estimator = tf.contrib.learn.SKCompat(tf.contrib.learn.DNNRegressor(model_dir='/tmp/DNN_offences_regression_trained_model', hidden_units=[20,18,14], optimizer=tf.train.AdamOptimizer(learning_rate=0.01), enable_centered_bias=False, feature_columns=tf.contrib.learn.infer_real_valued_columns_from_input(predictors.values)))

print("starting to train");

estimator.fit(predictors[:trainsize].values, targets[:trainsize].values.reshape(trainsize, noutputs)/scale_num_offences, steps=10000)

preds = estimator.predict(x=predictors[trainsize:].values)

predslistscale = preds['scores']*scale_num_offences

rmse = np.sqrt(np.mean((targets[trainsize:].values - predslistscale)**2))
print('DNNRegression has RMSE of {0}'.format(rmse));
```
